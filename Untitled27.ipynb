```python
class Queue:
  def __init__(self):
    self.queue=[]
    self.length_of_queue=10
  def enqueue(self):
    if(len(self.queue)!=10):
      print("\nEnter the task of the day: ")
      task=input()
      self.queue.append(task)
    else:
      print("\nToday's to-do list is full now!")
  def dequeue(self):
    if(len(self.queue)!=0):
      print("\nEnter the task number of the day you want to mark as done: ")
      task=int(input())
      if task < 1 or task > len(self.queue):
        print("No task exists at that number!")
      else:
        self.queue.pop(task - 1)
        print("Successfully marked as done.")

    else:
      print("\nToday's to-do list is already empty right now!")
  def peek(self):
    if(len(self.queue)==0):
      print("\nThere is no task right now for the day!")
    else:
      if(len(self.queue)==1):
        print("\n1. ",self.queue[0])
      else:
        i=1
        print('\n')
        for q in self.queue:
          print(i,". ",q)
          i+=1
  def delete(self):
    if(len(self.queue)==0):
      print("\nThere is no task to delete for the day!")
    else:
      del self.queue[:]
      print("\nYour tasks have been successfully deleted now.")
```

```python
print("------ TO-DO LIST APPLICATION (Using Queue) ------")
q=Queue()
check = 0

while True:
  print("\nEnter:\n1 for adding the task of the day.\n2 for removing a task of the day\n3 for checking the task of the day.\n4 for deleting all the tasks of the day.")
  user_task=input()
  if user_task=='1':
    q.enqueue()
  elif user_task=='2':
    q.dequeue()
  elif user_task=='3':
    q.peek()
  elif user_task=='4':
    q.delete()
  else:
    print("\nInvalid input please enter valid choices.")

  print("\nDo you want to continue? (y/n): ")
  question=input()
  if(question.lower()=='n'):
    print("Thank you for using the To-Do List Application!")
    break
  elif(question.lower()=='y'):
      continue
  else:
      print("\nInvalid choice please enter either 'y' for yes or 'n' for no: ")
  while question.lower()!='y' and question.lower()!='n':
    question=input()
    if(question.lower()=='n'):
      check=1
      break
    elif(question.lower()=='y'):
      check=0
      continue
    else:
      print("\nInvalid choice please enter either 'y' for yes or 'n' for no: ")
  if(check==1):
    print("\nThank you for using the To-Do List Application!")
    break
```
